```python
min_lat = 56.880525
max_lat = 57.073682
min_lon = 23.942295
max_lon = 24.255976

# Calculate the number of quadrants in each direction
num_quadrants_lat = math.ceil((max_lat - min_lat) / 0.005)
num_quadrants_lon = math.ceil((max_lon - min_lon) / 0.005)

# Create a map centered on Riga
map_riga = folium.Map(location=[(min_lat + max_lat) / 2, (min_lon + max_lon) / 2], zoom_start=11)
```

```python
map_riga
```

```python
quadrant_list = []
for lat in range(num_quadrants_lat):
    for lon in range(num_quadrants_lon):
        # Calculate the boundaries of the current quadrant
        quadrant_min_lat = min_lat + lat * 0.005
        quadrant_max_lat = quadrant_min_lat + 0.005
        quadrant_min_lon = min_lon + lon * 0.005
        quadrant_max_lon = quadrant_min_lon + 0.005
        quadrant_list.append((quadrant_min_lat, quadrant_max_lat, quadrant_min_lon, quadrant_max_lon))
        
```

```python
print(quadrant_list)
```

```python
for index, row in df.iterrows():
    lat = row['latitude']
    lon = row['longitude']
    for i, quadrant in enumerate(quadrant_list):
        if quadrant[0] <= lat < quadrant[1] and quadrant[2] <= lon < quadrant[3]:
            df.at[index, 'quadrant'] = i
            break
```

```python
from geopy import Nominatim

locator = Nominatim(user_agent='myGeocoder')
district_coords = []
for district, row in df_sell_grouped:
    print(district)
    location = locator.geocode(f"{district}, Riga")
    district_name = district
    if location:
        district_latitude = location.latitude
        district_longitude = location.longitude
    district_coords.append([district_name, district_latitude, district_longitude])
        
```

```python
district_coords
```

```python
import csv
with open('district_coords.csv', 'w', newline='', encoding='utf-8') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(['District', 'Latitude', 'Longitude'])  # Write header row
    writer.writerows(district_coords)
```

```python
min_lat = 56.880525
max_lat = 57.073682
min_lon = 23.942295
max_lon = 24.255976
map_object = folium.Map(location=[(min_lat + max_lat) / 2, (min_lon + max_lon) / 2], zoom_start=11)
for district in district_coords:
    latitude = district[1]
    longitude = district[2]
    marker = folium.Marker(location=[latitude, longitude])
    marker.add_to(map_object)
```
